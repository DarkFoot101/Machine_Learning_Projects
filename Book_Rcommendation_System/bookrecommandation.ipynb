```python
# To first create a popularity based system
ratings_with_name = ratings.merge(books, on = "ISBN")
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating' : 'num_ratings'}, inplace = True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating_df.rename(columns = {"Book-Rating" : "avg_ratings"}, inplace = True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on = "Book-Title")
popularity_df
```

```python
popularity_df = popularity_df[popularity_df["num_ratings"] >= 100].sort_values('avg_ratings', ascending= False).head(80)
```

```python
popularity_df = popularity_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings', 'Year-Of-Publication']]
```

```python
popularity_df.shape
```

```python
## now use the collabartive filter based approach
ratings_with_name
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 100
read_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(read_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 10
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns= 'User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0 , inplace = True)
```

```python
pt
```

```python
# these data are now similar, so we are going to take vectors of the books and check the similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores[0]
```

```python
def recommand(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x:x[1], reverse = True)[1:8]
    data =  []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].value))

        data.append(item)
    return data

    #index fetch
```
